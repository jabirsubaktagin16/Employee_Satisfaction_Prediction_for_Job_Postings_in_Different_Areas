```
class SurveyDataset(Dataset):
    def __init__(self, feature_list, label_list, transform=None):
        self.feature_list = feature_list
        self.label_list = label_list
        self.transform = transform


    def __len__(self):
        return self.feature_list.shape[0]

    def __getitem__(self, index):
        feature = self.feature_list[index]
        feature = feature.astype(np.float32)
        label = self.label_list[index]
        if self.transform is not None:
            feature = torch.tensor(feature)
         
        return feature, label-1
```

```
class NN_Prediction():
    def __init__(self):
        self.batch_size = 100
        self.num_iters = 3000
        self.loss_list = []
        self.input_dim = 3
        self.output_dim = 5
        self.dataloader_workers = 16
        self.model_optimizer = ''
        self.num_hidden = 100
        self.learning_rate = 0.001
        self.cv_loader = None
        self.model = None
        self.iteration_loss = 0
        self.dataset = None
        self.device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
        self.criterion = nn.CrossEntropyLoss()
        self.path = "/content.gdrive/My Drive/Dataset"

    def set_hyperparameters(self, batch, iters, dim, optimizer, hidden, lr):
        self.batch_size = batch
        self.num_iters = iters
        # always 9 for our case
        self.input_dim = dim
        self.num_hidden = hidden
        self.learning_rate = lr
        self.model_optimizer = optimizer
    
    def modify_dataset(self):
        self.dataset = pd.read_csv("/content/surveyDataset/Dataset/testing.csv")
        self.dataset["AGE_RANGE"] = self.dataset["AGE_RANGE"].replace(
            ["20-25", "26-30", "31-35", "36-40", "41 or Above"],
            [20, 26, 31, 36, 41])
        self.dataset["GENDER"] = self.dataset["GENDER"].replace(["Male", "Female", "Prefer not to disclose"], [1, 2, 3])
        self.dataset["OCCUPATION"] = self.dataset["OCCUPATION"].replace(
            ["Medical Field (Doctor, Nurse, Nutritionist, Pharmacists and other Health Care workers etc)",
             "Engineering and IT",
             "Business Field (Management, HR, Banking, Marketing etc)",
             "Academia (Teacher, Lecturer, Assistant/ Associate Professor, Professor)", "Student", "Unemployed",
             "Other"], [1, 2, 3, 4, 5, 6, 7])
        self.dataset["FIELD_OF_EDUCATION"] = self.dataset["FIELD_OF_EDUCATION"].replace(
            ["Medical, Biological or Chemical studies",
             "Engineering and IT",
             "Business Field", "Social Studies", "Other"],
            [1, 2, 3, 4, 5])
        self.dataset["MARITAL_STATUS"] = self.dataset["MARITAL_STATUS"].replace(["Married", "Unmarried"], [1, 2])
        self.dataset["SPOUSE_WILLING"] = self.dataset["SPOUSE_WILLING"].replace(
            ["Yes", "No", "My Spouse Does Not Work", "I am not Married"], [1, 2, 3, 4])
        self.dataset["SPOUSE_OCCUPATION"] = self.dataset["SPOUSE_OCCUPATION"].replace(
            ["Medical Field (Doctor, Nurse, Nutritionist, Pharmacists and other Health Care workers etc)",
             "Engineering and IT",
             "Business Field (Management, HR, Banking, Marketing etc)",
             "Academia (Teacher, Lecturer, Assistant/ Associate Professor, Professor)", "Student", "Unemployed",
             "Other",
             "I am not Married"], [1, 2, 3, 4, 5, 6, 7, 8])
        
    
    def write_dataset(self):
        #self.X = self.dataset[['SECURITY', 'SCHOOL', 'RENT', 'DISTANCE']].values
        self.y = self.dataset['ANSWER_1'].values
        self.X = self.dataset[['GENDER', 'AGE_RANGE', 'OCCUPATION', 'FIELD_OF_EDUCATION', 'HOMETOWN', 'MARITAL_STATUS', 'SPOUSE_WILLING',
                                'SPOUSE_OCCUPATION', 'PREF_RENT', 'PREF_HOMETOWN', 'PREF_SCHOOL', 'PREF_SPOUSE', 'PREF_SECURITY',
                                'SECURITY', 'SCHOOL', 'RENT', 'DISTANCE']].values

        X_train, X_cross_and_test, y_train, y_cross_and_test = train_test_split(self.X, self.y, test_size=0.4, random_state=0)
        x_cross, x_test, y_cross, y_test = train_test_split(X_cross_and_test, y_cross_and_test, test_size=0.5,
                                                            random_state=0)
        train_to_write = pd.DataFrame(X_train)
        train_to_write['ANSWER1'] = y_train
        train_to_write.to_csv(r'/content/surveyDataset/Dataset/training_data_whole.csv', index=False, header=True)
        cv_to_write = pd.DataFrame(x_cross)
        cv_to_write['ANSWER'] = y_cross
        cv_to_write.to_csv(
            r'/content/surveyDataset/Dataset/cross_validation_data_whole.csv',
            index=False, header= True)
        test_to_write = pd.DataFrame(x_test)
        test_to_write['ANSWER'] = y_test
        test_to_write.to_csv(r'/Users/aniquatabassum/Downloads/studies/Undergrad Thesis/SuverySetDivide/FinalWeights/testing_dataset_whole.csv',index=False, header= True)

    def def_training_input(self, features):
       
       input_features = self.dataset[[x for x in features]].values
       #print(input_features)
       return input_features
      

    def train_and_test(self, non_linearity, num_layers):
        iteration_loss = []
        self.dataset = pd.read_csv("/content/surveyDataset/Dataset/training_data_whole.csv")

        
        training_dataset = self.dataset[['GENDER', 'AGE_RANGE', "MARITAL_STATUS", 'SPOUSE_WILLING','OCCUPATION',  'SECURITY', 'SCHOOL', 'RENT', 'DISTANCE']].values
        training_dataset = np.array(training_dataset)[:, : self.input_dim]
        y_train = self.dataset['ANSWER1'].values
        y_train = np.array(y_train)
       
        self.dataset = pd.read_csv("/content/surveyDataset/Dataset/cross_validation_data_whole.csv")

        cv_datset = self.dataset[['GENDER', 'AGE_RANGE', "MARITAL_STATUS", 'SPOUSE_WILLING', 'OCCUPATION',  'SECURITY', 'SCHOOL', 'RENT',  'DISTANCE']].values
        cv_datset = np.array(cv_datset)[:, : self.input_dim]
        y_test = self.dataset['ANSWER'].values
        y_test = np.array(y_test)
        
        scaler = StandardScaler()
        training_dataset = scaler.fit_transform(training_dataset)
        cv_datset = scaler.transform(cv_datset)
        
        transformer = Normalizer()
        training_dataset = transformer.fit_transform(training_dataset)
        cv_datset = transformer.transform(cv_datset)
        
        training_survey_dataset = SurveyDataset(training_dataset, y_train, "tensor")

        cv_survey_dataset = SurveyDataset(cv_datset, y_test, "tensor")
       
        num_epochs = int(self.num_iters / (len(training_survey_dataset) / self.batch_size))
        

        train_loader = torch.utils.data.DataLoader(dataset= training_survey_dataset,
                                                   batch_size= self.batch_size,
                                                   shuffle=True,
                                                   num_workers= self.dataloader_workers)

        cv_loader = torch.utils.data.DataLoader(dataset=cv_survey_dataset,
                                                  batch_size= self.batch_size,
                                                  shuffle=False,
                                                  num_workers= self.dataloader_workers)


        '''
        INSTANTIATE MODEL CLASS
        '''
       
        if non_linearity == 'selu':
            if num_layers == 3:
                self.model = DeepNeuralNetworkModel(input_size=self.input_dim, num_classes=self.output_dim, num_hidden=self.num_hidden)
            elif num_layers == 5:
                self.model = DeepNeuralNetworkModel_SELU_5layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
        elif non_linearity == 'gelu':
            if num_layers == 3:
                self.model = DeepNeuralNetworkModel_Gelu(input_size=self.input_dim,
                                                    num_classes=self.output_dim,
                                                    num_hidden=self.num_hidden)
            elif num_layers == 5:
                self.model = DeepNeuralNetworkModel_Gelu_5layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
            elif num_layers == 7:
                self.model = DeepNeuralNetworkModel_Gelu_7layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
        elif non_linearity == 'relu':
            if num_layers == 3:
                self.model = DeepNeuralNetworkModel_RELU_3layer(input_size=self.input_dim,
                                                    num_classes=self.output_dim,
                                                    num_hidden=self.num_hidden)
            elif num_layers == 5:
                self.model = DeepNeuralNetworkModel_RELU_5layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
            elif num_layers == 7:
                self.model = DeepNeuralNetworkModel_RELU_7layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
        
        elif non_linearity == 'relu6':
            if num_layers == 3:
                self.model = DeepNeuralNetworkModel_RELU6_3layer(input_size=self.input_dim,
                                                    num_classes=self.output_dim,
                                                    num_hidden=self.num_hidden)
            elif num_layers == 5:
                self.model = DeepNeuralNetworkModel_RELU6_5layer(input_size=self.input_dim,
                                                           num_classes=self.output_dim,
                                                           num_hidden=self.num_hidden)
            

        # To enable GPU
        self.model.to(self.device)
        # INSTANTIATE OPTIMIZER CLASS
        #optimizer = None
        if self.model_optimizer == 'sgd':
            self.model_optimizer = torch.optim.SGD(self.model.parameters(), lr=self.learning_rate)
        elif self.model_optimizer == 'adam':
            self.model_optimizer = torch.optim.Adam(self.model.parameters(), lr=self.learning_rate)
        elif self.model_optimizer == 'adargrad':
            self.model_optimizer = torch.optim.Adagrad(self.model.parameters(), lr=self.learning_rate)

        '''
        TRAIN THE MODEL
        '''

        iter = 0
        loss_value = 0
        accuracy_value = 0
        for epoch in range(num_epochs):
            for i, (person_info, labels) in enumerate(train_loader):

                person_info = person_info.to(self.device)
                labels = labels.to(self.device)

                # Clear gradients w.r.t. parameters
                self.model_optimizer.zero_grad()

                # Forward pass to get output/logits
                outputs = self.model(person_info)
                # Calculate Loss: PyTorch implementation of CrossEntropyLoss works with logits, not probabilities
                loss = self.criterion(outputs, labels)
                # Getting gradients w.r.t. parameters
                loss.backward()

                # Updating parameters
                self.model_optimizer.step()
                
                iter += 1

                if iter % 500 == 0:
                    # Calculate Accuracy
                    correct = 0
                    total = 0
                    # Iterate through test dataset
                    for person_info, labels in cv_loader:

                        person_info = person_info.to(self.device)

                        # Forward pass only to get logits/output
                        outputs = self.model(person_info)

                        # Get predictions from the maximum value
                        _, predicted = torch.max(outputs, 1)

                        # Total number of labels
                        total += labels.size(0)

                        # Total correct predictions
                        if torch.cuda.is_available():
                            correct += (predicted.cpu() == labels.cpu()).sum()
                        else:
                            correct += (predicted == labels).sum()

                    accuracy = 100 * correct.item() / total

                    # Print Loss
                    self.loss_list.append(loss.item())
                    print('Iteration: {}. Loss: {}. Accuracy: {}'.format(iter, loss.item(), accuracy))
                    u = (predicted.cpu() - label.cpu()).sum()
                    v = (label.cpu() - label.cpu().mean()).sum()
                    loss_value = 1- (u/v)
                    accuracy_value = accuracy

        print('Iteration: {}. Loss: {}. Accuracy: {}'.format(iter, loss_value, accuracy_value))


    def show_loss_graph(self):
        import matplotlib
        import matplotlib.pyplot as plt

        print(self.loss_list)
        plt.plot(self.loss_list)
        plt.ylabel('Cross Entropy Loss')
        plt.xlabel('Iteration (in every 500)')
        plt.show()
    
    def save_model(self, name):
        save_model = True

        if save_model is True:
           torch.save(self.model.state_dict(), self.path + "/"+name + '.pkl') 

    def predict_output(self):
        total = 0 
        correct = 0 
        prdicted_list = []
        correct_list = []
        self.dataset = pd.read_csv("/content/surveyDataset/Dataset/testing_dataset_whole.csv")

        test_dataset = self.dataset[['GENDER', 'AGE_RANGE', "MARITAL_STATUS", 'SPOUSE_WILLING', 'OCCUPATION',  'SECURITY', 'SCHOOL', 'RENT',  'DISTANCE']].values
        test_dataset = np.array(test_dataset)[:, : self.input_dim]
        y_test = self.dataset['ANSWER'].values
        y_test = np.array(y_test)
        test_survey_dataset = SurveyDataset(test_dataset, y_test, "tensor")
        test_loader = torch.utils.data.DataLoader(dataset=test_survey_dataset,
                                                  batch_size= self.batch_size,
                                                  shuffle=False,
                                                  num_workers= self.dataloader_workers)
        i = 1
        for person_info, labels in test_loader:

                        person_info = person_info.to(self.device)

                        # Forward pass only to get logits/output
                        outputs = self.model(person_info)

                        # Get predictions from the maximum value
                        _, predicted = torch.max(outputs, 1)

                        # Total number of labels
                        total += labels.size(0)

                        # Total correct predictions
                        if torch.cuda.is_available():
                            correct += (predicted.cpu() == labels.cpu()).sum()
                        else:
                            correct += (predicted == labels).sum()
                        print("number: ", i)
                        i+=1
                        #prdicted_list.extend(predicted.data.numpy().tolist())
                        #correct_list.extend(labels.data.numpy().tolist())
                        
        accuracy = 100 * correct.item() / total
        #print(prdicted_list)
        #print(correct_list)
        #preditcted_vs_actual = zip(prdicted_list, correct_list)
        #print("predicted  Actual")
        #for i in range(len(prdicted_list)):
          #print(prdicted_list[i], "  ", correct_list[i])
        print("CV accuracy is ", accuracy)


    

```
